```python
# Correlation
import pandas as pd
import numpy as np
url = 'Salary_Data.csv'
data = pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()
plt.show()
correlation = data.corr()
print(correlation)
import seaborn as sns
sns.heatmap(correlation, annot = True)
data.plot(kind = 'scatter', x = 'Salary', y = 'YearsExperience', figsize = (9,9), color = 'black')
#mpg phụ thuộc vào cái lớn nhất
from sklearn import linear_model
# Hồi quy tuyến tính bậc 1
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['Salary']), y = data['YearsExperience'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác', model.score(X = pd.DataFrame(data['Salary']), y = data['YearsExperience']))
# pt hồi quy: y = coef*X + intercept
#data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black', xdim = (0,7))
y_pred = model.predict(X = pd.DataFrame(data['Salary']))
plt.plot(data['Salary'], y_pred, color = 'blue')
# Hồi quy tuyến tính bậc 2
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Salary'], data['Salary']**2]).T
model.fit(X, y = data['YearsExperience'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác bậc 2', model.score(X, y = data['YearsExperience']))
range = np.arange(0, 120000, 1000)
data.plot(kind = 'scatter', x = 'Salary', y = 'YearsExperience', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict( X = predictor)
plt.plot( range, y_pred, color = 'blue')
# Hồi quy tuyến tính bậc 5
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Salary'], data['Salary']**2, data['Salary']**3,
                  data['Salary']**4, data['Salary']**5]).T
model.fit(X, y = data['YearsExperience'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác bậc 5', model.score(X, y = data['YearsExperience']))
range = np.arange(0, 120000, 1000)
data.plot(kind = 'scatter', x = 'Salary', y = 'YearsExperience', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range,range**2, range**3, range**4, range**5]).T
y_pred = model.predict( X = predictor)
plt.plot( range, y_pred, color = 'blue')


# Hồi quy tuyến tính bậc 10
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Salary'], data['Salary']**2, data['Salary']**3, 
                  data['Salary']**4, data['Salary']**5, data['Salary']**6, 
                  data['Salary']**7, data['Salary']**8, data['Salary']**9 ,data['Salary']**10]).T
model.fit(X, y = data['YearsExperience'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác bậc 10', model.score(X, y = data['YearsExperience']))
range = np.arange(0, 120000, 1000)
data.plot(kind = 'scatter', x = 'Salary', y = 'YearsExperience', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range,range**2, range**3, range**4, range**5, 
                          range**6, range**7, range**8, range**9, range**10]).T
y_pred = model.predict( X = predictor)
plt.plot( range, y_pred, color = 'blue')
```
